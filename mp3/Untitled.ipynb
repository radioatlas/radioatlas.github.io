```python
from pytube import Playlist
from datetime import datetime
import scrapetube

# importing packages 
from pytube import YouTube 
import os 
```

```python
def videos_urls_channel(chan_url):
    videos = scrapetube.get_channel(chan_url)
    VIDEOS=[]
    for video in videos:
        VIDEOS.append('https://www.youtube.com/watch?v='+video['videoId'])
    return VIDEOS
URLS= videos_urls_channel("UCPsOKetsJXoAtenefkwp85w")
```

```python
def download_list_videos(list_videos):
    log_name='log_'+datetime.today().strftime('%Y_%m_%d_%H_%M_%S')+'.txt'
    N=len(list_videos)
    with open(log_name, 'w') as f:
        f.close()
    for i,x in enumerate(list_videos):
        try:  
            yt = YouTube( x)
            # extract only audio 
            video = yt.streams.filter(only_audio=True).first() 
            out_file = video.download()
            #out_file=out_file.split('\\')[-1]

            try:
                # save the file 
                base, ext = os.path.splitext(out_file) 
                new_file = datetime.today().strftime('%Y_%m_%d_%H_%M_%S') + '.mp4'
                os.rename(out_file, new_file) 

                # result of success 
                print(i,N,out_file + '<;>'+new_file+'<;>>>>'+ "OK ")
                print(out_file + '<;>'+new_file+'<;>>>>'+ "OK ", file=open(log_name, 'a', encoding='utf-8'))
            except:
                print(x+ " >>> OK RENAME NOK")
                print(x+ " >>> OK RENAME NOK", file=open(log_name, 'a', encoding='utf-8'))

        except Exception as e:
            print(e)
            print(x+ " >>> ERROOOOOOOOOR")
            print(x+ " >>> ERROOOOOOOOOR", file=open(log_name, 'a', encoding='utf-8'))
```

```python
URLS= videos_urls_channel("UCPsOKetsJXoAtenefkwp85w")
download_list_videos(list_videos=URLS)
```
